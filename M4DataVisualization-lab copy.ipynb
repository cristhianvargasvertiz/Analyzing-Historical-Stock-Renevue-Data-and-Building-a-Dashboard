Recordar que cuando agrupas creas un objeto o Serie de Pandas y deja de ser un DataFrame, una de la razones es que cuando agrupas el resultado, por tanto si quieres acceder al index de "top_5" puedes hacerlo a traves de .iloc[].index

Esto lo necesitaras para crear los Labels.

```python
QUERY= """

SELECT *
FROM DatabaseDesireNextYear

"""

df = pd.read_sql(QUERY, conn)

group = df.groupby("DatabaseDesireNextYear")["Respondent"].sum()
top_5 = group.sort_values(ascending=False).head(5)

label_X = top_5.iloc[0:5].index

# otra opcion: label_X = top_5.iloc[0:5].reset_index()['DatabaseDesireNextYear']
# para recordar usando reset_index puedes volver un objeto en un Dataframe implementando solo index verticales
# los titulos o etiquetas de las columnas no son indexadas
# este caso top_5 se volvio una serie pandas u objeto con esta estructura por que fue el resultado de una agrupacion: 
# ([53804386, 45897489, 41370718, 41353217, 35314757], 
# index=['PostgreSQL', 'MongoDB', 'Redis', 'MySQL', 'Elasticsearch'])
# por eso se accedio a los datos a traves de la funcion .index
# pero tambien se puedo hacer a traves de .reset_index() que vuelve objetos en Dataframe, como se muestra en la primera linea

plt.figure(figsize=(15, 15))

fig, ax = plt.subplots()
ax.pie(top_5, labels=label_X, autopct='%1.1f%%', explode = (0.1, 0, 0, 0, 0), shadow = True)
ax.set_title("Data bases Desire Next Year")
plt.show()
```

```python
QUERY = """

SELECT *
FROM DatabaseWorkedWith

"""
df = pd.read_sql_query(QUERY,conn)

# Supongamos que df es tu DataFrame con las columnas Respondent y LanguageDesireNextYear

# Encuentra los Respondent que aparecen solo una vez
unique_respondents = df['Respondent'].value_counts()[df['Respondent'].value_counts() == 1].index

# Filtra los datos para aquellos Respondent que también tienen 'MySQL' como LanguageDesireNextYear
mysql_only_respondents = df[(df['Respondent'].isin(unique_respondents)) & (df['DatabaseWorkedWith'] == 'MySQL')]

# Imprime los Respondent que cumplen ambas condiciones
len(mysql_only_respondents['Respondent'])
```

SELECT DISTINCT selecionara todos valores que aparecen muchas veces REPRESENTADOS en un valor, es decir que eliminara duplicados basado en Respondent, en el caso que se haga un WHERE en esa instancia con MySQL, solo devolvera los valores unicos REPRESENTATIVOS sin duplicados pero no se sabra si esos habran aparecido nuevamente, es decir que si el ID 4 aparecio dos veces, se mostrara cumpliendo las clausulas, pero lo que nosotros queremos es filtrar Repondents que solo hayan aparecido una vez, por eso debemos interceptarlos con having count(*) = 1 , para que devuelva solo los que tienen una aparacion en Respondent con el condicional de que esten trabajando actualemente con MySQL en la columna DatabaseWorkedWith.

DatabaseWorkedWith se guarda como t1 y se puede invocar a sus columnas como t1.Respondent o t1.DatabaseWorkedWith
y t2 es la tabla resultante de la consulta que agrupa la fila Respondent pero selecciona sola los que se cuentan una vez.

Despues puedes hacer la interseccion y ponerle la clausula de solo los que tengan los valores ="MySQL"


```python
QUERY = """

SELECT DISTINCT t1.Respondent 
FROM DatabaseWorkedWith AS t1

INNER JOIN (
    SELECT Respondent
    FROM DatabaseWorkedWith
    GROUP BY Respondent
    HAVING COUNT(*) = 1) AS t2 
    
    ON t1.Respondent = t2.Respondent

WHERE t1.DatabaseWorkedWith = 'MySQL';

"""
df = pd.read_sql_query(QUERY,conn)

df
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """

SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35

"""
df = pd.read_sql_query(QUERY,conn)

work_week = df['WorkWeekHrs'].quantile(0.5)
code_rev = df['CodeRevHrs'].quantile(0.5)


plt.figure(figsize=(15, 15))

fig, ax = plt.subplots()

ax.bar('Age 30 - 35', work_week, label='WorkWeekHrs')
ax.bar('Age 30 - 35', code_rev, bottom=work_week, label='CodeRevHrs')
ax.legend()

plt.show()
```

### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """

SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60

"""
df = pd.read_sql_query(QUERY,conn)

medians = df.groupby('Age')['ConvertedComp'].median()

data = pd.DataFrame(list(zip(medians.index, medians.values)),
               columns =['Age', 'MedConvertedComp'])

plt.rcParams['figure.figsize'] = [8,8]

sns.lineplot(data=data, x='Age', y = 'MedConvertedComp')

plt.show()
```
